```python
# import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Excel file

df = pd.read_csv('Sample - Superstore.csv')

df
```

```python
# Basic Overview
print("Shape of the dataset:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
print("\nSample Rows:\n", df.head())
```

```python
# Clean column names (remove extra spaces)
df.columns = df.columns.str.strip()
```

```python
# Summary Statistics
print("\nSummary Statistics:\n", df.describe())
```

```python
# Univariate Analysis

sns.set_style("darkgrid")

numerical_columns = df.select_dtypes(include=["int64", "float64"]).columns

plt.figure(figsize=(14, len(numerical_columns) * 3))
for idx, feature in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 2, idx)
    sns.histplot(df[feature], kde=True)
    plt.title(f"{feature} | Skewness: {round(df[feature].skew(), 2)}")

plt.tight_layout()
plt.show()
```

```python
#Product Performance

#  Which product sub-categories have the highest sales volume?
sales_by_subcategory = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False)

#  Are there underperforming products with negative or very low profits?
profit_by_product = df.groupby("Product Name")["Profit"].sum().sort_values()
underperforming_products = profit_by_product[profit_by_product < 0].head(10)

#  Correlation between Quantity, Sales, and Profit
correlation_matrix = df[["Quantity", "Sales", "Profit"]].corr()

# Plotting
plt.figure(figsize=(6, 3))
sales_by_subcategory.plot(kind="bar", color="skyblue", title="Total Sales by Sub-Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 3))
underperforming_products.plot(kind="barh", color="salmon", title="Top 10 Underperforming Products (Negative Profit)")
plt.xlabel("Total Profit")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 3))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation: Quantity, Sales, Profit")
plt.tight_layout()
plt.show()
```

```python
#Geographic Trends

# Top 10 states by revenue
revenue_by_state = df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10)

# States with negative profit (geographic areas losing money)
loss_by_state = df.groupby("State")["Profit"].sum().sort_values()
losing_states = loss_by_state[loss_by_state < 0]

# Sales distribution by Shipping Mode and Region
sales_by_ship_region = df.groupby(["Region", "Ship Mode"])["Sales"].sum().unstack().fillna(0)

# Plot top revenue-generating states
plt.figure(figsize=(6, 3))
revenue_by_state.plot(kind="bar", color="royalblue", title="Top 10 States by Revenue")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot states with negative profit
plt.figure(figsize=(6, 3))
losing_states.plot(kind="bar", color="tomato", title="States with Negative Profit")
plt.ylabel("Total Profit")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Heatmap for sales by shipping mode and region
plt.figure(figsize=(5, 3))
sns.heatmap(sales_by_ship_region, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Sales Distribution by Ship Mode and Region")
plt.tight_layout()
plt.show()

# Return data for reference
revenue_by_state, losing_states, sales_by_ship_region
```

```python
# Time-Based Trends

# Convert date column and extract Year and Month
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month

# Yearly Sales and Profit
yearly_data = df.groupby('Year')[['Sales', 'Profit']].sum()

#  Monthly Sales Trends
monthly_sales = df.groupby('Month')['Sales'].sum()

# Sales by Month and Category
monthly_category_sales = df.groupby(['Month', 'Category'])['Sales'].sum().unstack().fillna(0)

# Plot  Yearly Sales and Profit
plt.figure(figsize=(5, 3))
yearly_data.plot(kind="line",marker='o')
plt.title("Yearly Sales and Profit")
plt.ylabel("Sales/Profit")
plt.tight_layout()
plt.show()

# Plot Monthly Sales Trends
plt.figure(figsize=(5, 3))
monthly_sales.plot(kind="line", marker='o', color="orange")
plt.title("Monthly Sales Trend (All Years Combined)")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot pie of Sales by Month and Category
plt.figure(figsize=(10, 8))
monthly_category_sales.plot(kind="pie", subplots=True, figsize=(10, 8), autopct='%1.1f%%')
plt.tight_layout()
plt.show()

```
