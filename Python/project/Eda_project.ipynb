```python
# import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Excel file

df = pd.read_csv('Sample - Superstore.csv')

df
```

```python
# Basic Overview
print("Shape of the dataset:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
print("\nSample Rows:\n", df.head())
```

```python
# Clean column names (remove extra spaces)
df.columns = df.columns.str.strip()
```

```python
# Summary Statistics
print("\nSummary Statistics:\n", df.describe())
```

```python
# Univariate Analysis

sns.set_style("darkgrid")

numerical_columns = df.select_dtypes(include=["int64", "float64"]).columns

plt.figure(figsize=(14, len(numerical_columns) * 3))
for idx, feature in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 2, idx)
    sns.histplot(df[feature], kde=True)
    plt.title(f"{feature} | Skewness: {round(df[feature].skew(), 2)}")

plt.tight_layout()
plt.show()
```

```python
# Categorical Counts
plt.figure(figsize=(8,4))
sns.countplot(data=df, x='Category')
plt.title('Product Category Count')
plt.show()
```

```python
# most sales in Region pie chart
plt.figure(figsize=(8,4))
df['Region'].value_counts().plot(kind='pie', autopct='%1.1f%%')

sns.countplot(data=df, x='Region')
plt.title('Region Sales Distribution')
plt.ylabel('')
plt.show()
```

```python
from pickle import TRUE
## Clean column names
df.columns = df.columns.str.replace('ï»¿', '')

# Convert 'Order Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

# Define 'Recession' and 'Non-Recession' periods based on year
df['Year'] = df['Order Date'].dt.year
df['Period'] = df['Year'].apply(lambda x: 'Recession' if x == 2020 else 'Non-Recession')

# Group by Category and Period to get total Sales
sales_comparison = df.groupby(['Period', 'Category'])['Sales'].sum().reset_index()

# Pivot table for better plotting
sales_comparison_pivot = sales_comparison.pivot(index='Category', columns='Period', values='Sales')

# Plot the comparison
sales_comparison_pivot.plot(kind='bar', figsize=(6, 3), color=['#4c72b0', '#dd8452'])

plt.title('Sales Comparison by Category (Recession vs Non-Recession)')
plt.ylabel('Total Sales')
plt.xlabel('Category')
plt.xticks(rotation=1)
plt.legend(title='Period',loc='upper right',bbox_to_anchor=(0.95, 0.95))
plt.tight_layout()
plt.show()
```

```python
''' Plot different lines for categories of Category and
analyze the trend to answer the question Is there a noticeable difference in
sales trends between different Category during recession periods?'''

# Assuming 'Recession_Period' is a boolean indicating recession periods (True/False)
df['Recession_Period'] = df['Year'].apply(lambda x: 'Recession' if x == 2020 else 'Non-Recession')

# Convert 'Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Plotting sales trends for each Sub-Category
plt.figure(figsize=(12, 8))

# Loop through each Sub-Category and plot the sales data
for category in df['Category'].unique():
    category_data = df[df['Category'] == category]
    # Plot the sales, and differentiate recession and non-recession periods
    recession_data = category_data[category_data['Recession_Period'] == 'Recession']
    non_recession_data = category_data[category_data['Recession_Period'] == 'Non-Recession']

    plt.plot(recession_data['Order Date'], recession_data['Sales'], label=f'{category} (Recession)', marker='o')
    plt.plot(non_recession_data['Order Date'], non_recession_data['Sales'], label=f'{category} (Non-Recession)', marker='o')

# Adding labels, title, and legend
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales Trends by Category (Recession vs Non-Recession)')
plt.legend()
plt.grid(True)

# shoe plot
plt.show()
```
